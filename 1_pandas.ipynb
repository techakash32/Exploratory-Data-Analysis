```python
import plotly.graph_objects as go
import plotly.io as pio
pio.templates.default="plotly_white"
```

```python
print(df.isnull().sum())
```

```python
#check persentages of attrition by department
```

```python
# filter the data to show only "YES" values in the "ATTRITION" coloumn
```

```python
attr_df=df[df['Attrition']=='Yes']
```

```python
attr_df
```

```python
attrition_by_dpt=attr_df.groupby(['Department']).size().reset_index(name="count")
attrition_by_dpt
```

```python
#create a donut chart
fig=go.Figure(data=[go.Pie(
    labels=attrition_by_dpt['Department'],
    values=attrition_by_dpt['count'],
    hole=0.4,
    marker=dict(colors=['red','green']),
    textposition='inside'
)])
#to update layout
fig.update_layout(title='ATTRITION BY DEPARTMENT')

#show the chart
fig.show()
```

```python
attr_df=df[df['Attrition']=='Yes']
attrition_by_edu=attr_df.groupby(['EducationField']).size().reset_index(name="count")
attrition_by_edu
```

```python
#create a donut chart
fig=go.Figure(data=[go.Pie(
    labels=attrition_by_edu['EducationField'],
    values=attrition_by_edu['count'],
    hole=0.4,
    marker=dict(colors=['red','green']),
    textposition='inside'
)])
#to update layout
fig.update_layout(title='ATTRITION BY EducationField')

#show the chart
fig.show()
```

```python
attrition_by=attr_df.groupby(['YearsAtCompany']).size().reset_index(name="count")
attrition_by
```

```python
attrition_by=attr_df.groupby(['YearsSinceLastPromotion']).size().reset_index(name="count")
attrition_by
```

```python
attrition_by=attr_df.groupby(['Gender']).size().reset_index(name="count")
attrition_by
```

```python
#create a donut chart
fig=go.Figure(data=[go.Pie(
    labels=attrition_by['Gender'],
    values=attrition_by['count'],
    hole=0.4,
    marker=dict(colors=['red','green']),
    textposition='inside'
)])
#to update layout
fig.update_layout(title='ATTRITION BY GENDER')

#show the chart
fig.show()
```
